```python
def add_features(x_train):
    # shift time,uin-after,u-out-after by one by default 
    # shift by one for x_train,x_test
    # shift time,uin-after,u-out-after by one by default 
    # shift by one for x_train,x_test
    x_train['time_after'] = data[['time_step_std']].shift(1)
    x_train['uin_after'] = data[['un_in_std']].shift(1)
    x_train['u_out_after'] = data[['u_out']].shift(1)
    x_train['time_after2'] = data[['time_step_std']].shift(2)
    x_train['uin_after2'] = data[['un_in_std']].shift(2)
    x_train['u_out_after2'] = data[['u_out']].shift(2)
    x_train['time_after3'] = data[['time_step_std']].shift(3)
    x_train['uin_after3'] = data[['un_in_std']].shift(3)
    x_train['u_out_after3'] = data[['u_out']].shift(3)
    x_train['time_after4'] = data[['time_step_std']].shift(4)
    x_train['uin_after4'] = data[['un_in_std']].shift(4)
    x_train['u_out_after4'] = data[['u_out']].shift(4)
    x_train['time_after5'] = data[['time_step_std']].shift(5)
    x_train['uin_after5'] = data[['un_in_std']].shift(5)
    x_train['u_out_after5'] = data[['u_out']].shift(5)   
    x_train['time_step_back'] = data[['time_step_std']].shift(-1)
    x_train['u_in_back'] = data[['un_in_std']].shift(-1)
    x_train['uout_back'] = data[['u_out']].shift(-1)
    x_train['time_step_back2'] = data[['time_step_std']].shift(-2)
    x_train['u_in_back2'] = data[['un_in_std']].shift(-2)
    x_train['uout_back2'] = data[['u_out']].shift(-2)
    x_train['time_step_back3'] = data[['time_step_std']].shift(-3)
    x_train['u_in_back3'] = data[['un_in_std']].shift(-3)
    x_train['uout_back3'] = data[['u_out']].shift(-3)
    x_train['time_step_back4'] = data[['time_step_std']].shift(-4)
    x_train['u_in_bac4'] = data[['un_in_std']].shift(-4)
    x_train['uout_back4'] = data[['u_out']].shift(-4)
    x_train['time_step_back5'] = data[['time_step_std']].shift(-5)
    x_train['u_in_bac5'] = data[['un_in_std']].shift(-5)
    x_train['uout_back5'] = data[['u_out']].shift(-5)
    print("step 1- completed")
    #The thinking behind this is that the pressure in the lungs is proportional to 
    #how much air has actually been pumped into them. Obviously this feature is not useful when breathing out,
    #but as the expiratory phase is not scored we are ok.
    #In other words, the pressure would be correlated to the integral of valve opening
    
    x_train['u_in_cumsum'] = data[['u_in']].cumsum()
    x_train['time_step_cumsum'] = data[['time_step_std']].cumsum()
    x_train['last_value_u_in'] = data[['u_in']].transform('last')
    x_train['breath_id__u_in__max'] = data[['u_in']].transform('max')
    x_train['breath_id__u_out__max'] =data[['u_out']].transform('max')
    x_train['breath_id__u_in__mean'] =data[['u_in']].transform('mean')
    x_train['breath_id__u_in__diffmax'] = x_train.groupby(['breath_id'])['u_in'].transform('max') - x_train['u_in']
    x_train['breath_id__u_in__diffmean'] = x_train.groupby(['breath_id'])['u_in'].transform('mean') - x_train['u_in']
    print("Step-2...Completed")
    
    # difference between consequitive values
    x_train['u_in_diff1'] = x_train['u_in'] - x_train['uin_after']
    x_train['u_out_diff1'] = x_train['u_out']- x_train['u_out_after']
    x_train['u_in_diff2'] = x_train['u_in'] - x_train['uin_after2']
    x_train['u_out_diff2'] = x_train['u_out'] - x_train['u_out_after2']
    print("Step-3...Completed")
    # dummy vARIABLES  
    x_train['R'] = x_train['R'].astype(str)
    x_train['C'] = x_train['C'].astype(str)
    x_train['R__C'] =  x_train["R"].astype(str) + '__' +  x_train["C"].astype(str)
    print("Step-4...Completed")
    x_train['cross']= x_train['u_in'] * x_train['u_out']
    x_train['cross2']= x_train['time_step'] * x_train['u_out']
    x_train['area'] = x_train['time_step'] * x_train['u_in']
    x_train['area'] = x_train.groupby('breath_id')['area'].cumsum()
    x_train['time_step_cumsum'] = x_train.groupby(['breath_id'])['time_step'].cumsum()
    x_train['u_in_cumsum'] = (x_train['u_in']).groupby(x_train['breath_id']).cumsum()
    print("Step-5...Completed")
    
    # ewm mean
    x_train['time_step_diff'] = x_train.groupby('breath_id')['time_step'].diff().fillna(0)
    x_train['ewm_u_in_mean'] = (x_train.groupby('breath_id')['u_in'].ewm(halflife=9).mean().reset_index(level=0,drop=True))
    # rolling window 
    x_train[["15_in_sum","15_in_min","15_in_max","15_in_mean"]] = (x_train.groupby('breath_id')['u_in'].rolling(window=15,min_periods=1).agg({"15_in_sum":"sum", "15_in_min":"min",  "15_in_max":"max","15_in_mean":"mean"}).reset_index(level=0,drop=True))
    print("Step-6...Completed")
    x_train = pd.get_dummies(x_train)
    x_train['u_in_lagback_diff1'] = x_train['u_in'] - x_train['u_in_back']
    x_train['u_out_lagback_diff1'] = x_train['u_out'] - x_train['uout_back']
    x_train['u_in_lagback_diff2'] = x_train['u_in'] - x_train['u_in_back2']
    x_train['u_out_lagback_diff2'] = x_train['u_out'] - x_train['uout_back2']
    x_train=x_train.fillna(0)

    print("Step-7...Completed")
    
    
    return x_train
```

```python
# fix nans in y_train 
# Filling NaN in Categorical data
#1. Function to replace NAN values with mode value
def impute_nan_most_frequent_category(ColName):
    # .mode()[0] - gives first category name
     most_frequent_category=ColName.mode()[0]
    
    # replace nan values with most occured category
     
     ColName.fillna(most_frequent_category,inplace=True)
     return ColName

    
```
